## Part 1: Lists

A **list** is an ordered collection of items. Think of it like a roster, a playlist, or a grocery list.

```python
my_list = [item1, item2, item3]
```

Lists can hold any type of data: strings, numbers, even other lists.

### 1.1 Creating and Exploring Lists

Imagine you're a teacher tracking student scores on a recent quiz.

````python
# Quiz scores for 8 students (out of 100)
scores = [88, 92, 75, 61, 95, 83, 70, 100]

print("All scores:", scores)
print("First score:", scores[0])
print("Last score:", scores[-1])
print("Number of students:", len(scores))
````

### 1.2 Processing Lists with Loops

Loops let us do something with **every** item in a list. This is where lists become powerful — instead of writing repetitive code, we write the logic once and the loop handles the rest.

````python
# Print each score with a label
for i in range(len(scores)):
    print(f"Student {i + 1}: {scores[i]}")
````

````python
# Calculate the class average
total = 0
for score in scores:
    total += score

average = total / len(scores)
print(f"Class average: {average:.1f}")
````

````python
# Find all students who scored 90 or above ("A" grades)
a_grades = []
for score in scores:
    if score >= 90:
        a_grades.append(score)

print(f"{len(a_grades)} students earned an A: {a_grades}")
````

### 1.4 Lists from the Real World: Fetching Data from an API

An **API** (Application Programming Interface) lets your code request data from a server on the internet. We'll use Python's built-in `urllib` module — no extra installs or API keys needed.

We'll pull data about **countries** from the REST Countries API.

````python
import urllib.request
from urllib.parse import urlencode
import json

# Fetch data about countries in South America
url = "https://restcountries.com/v3.1/region/south%20america"
params = {
    "fields": "name,population,area"
}
response = urllib.request.urlopen(f"{url}?{urlencode(params)}")
countries = json.loads(response.read())

print(f"Got data for {len(countries)} countries!")
print("Here's what one entry looks like:")
print(json.dumps(countries[0], indent=2))
````

````python
# Build a list of just the country names
names = []
for country in countries:
    names.append(country["name"]["common"])

print("South American countries:")
for name in sorted(names):
    print(f"  - {name}")
````

````python
# Build a list of populations and find the total
populations = []
for country in countries:
    populations.append(country["population"])

total_pop = 0
for pop in populations:
    total_pop += pop

print(f"Total population of South America: {total_pop:,}")
````

## Part 2: Dictionaries

A **dictionary** stores data as **key-value pairs**. Instead of accessing items by position (like a list), you access them by a meaningful label.

```python
my_dict = {"key": "value", "name": "Alice", "age": 17}
```

Dictionaries are perfect when each piece of data has a label — like a student record, a product, or a weather report.

### 2.1 Creating and Exploring Dictionaries

Imagine you're building a simple student record system.

````python
# A single student record
student = {
    "name": "Maria Santos",
    "grade": 11,
    "gpa": 3.7,
    "courses": ["AP Chemistry", "English 11", "Precalculus", "US History"]
}

print("Name:", student["name"])
print("GPA:", student["gpa"])
print("Number of courses:", len(student["courses"]))
````

````python
# Loop through the keys and values
print("--- Full Student Record ---")
for key, value in student.items():
    print(f"  {key}: {value}")
````

### 2.2 Lists of Dictionaries

In the real world, you usually have **many** records. That means a **list of dictionaries** — each dictionary is one record.

````python
students = [
    {"name": "Maria Santos", "gpa": 3.7, "absences": 2},
    {"name": "James Lee", "gpa": 3.2, "absences": 8},
    {"name": "Aisha Patel", "gpa": 3.9, "absences": 0},
    {"name": "Carlos Rivera", "gpa": 2.8, "absences": 5},
    {"name": "Lily Chen", "gpa": 3.5, "absences": 1},
]

# Print a report card summary for each student
for s in students:
    print(f"{s['name']} — GPA: {s['gpa']}, Absences: {s['absences']}")
````

````python
# Find students with perfect attendance
perfect = []
for s in students:
    if s["absences"] == 0:
        perfect.append(s["name"])

print("Perfect attendance:", perfect)
````

````python
# Calculate the average GPA
total_gpa = 0
for s in students:
    total_gpa += s["gpa"]

avg_gpa = total_gpa / len(students)
print(f"Average GPA: {avg_gpa:.2f}")
````

### 2.4 Dictionaries from the Real World: University API

Let's fetch real university data and process it as dictionaries. We'll use the Hipolabs Universities API — free, no key needed.

````python
import urllib.request
from urllib.parse import urlencode
import json

# Fetch data about countries in South America
url = "http://universities.hipolabs.com/search"
params = {
    "name": "technology",
    "country":"United+States"
}
response = urllib.request.urlopen(f"{url}?{urlencode(params, safe='+')}")
universities = json.loads(response.read())

print(f"Found {len(universities)} universities!")
print("\nHere's what one record looks like:")
print(json.dumps(universities[0], indent=2))
````

````python
# Each university is a dictionary. Let's explore the keys.
print("Keys in each record:", list(universities[0].keys()))
````

````python
# List all university names
print("Technology universities in the US:\n")
for uni in universities:
    print(f"  {uni['name']}")
````

````python
# Build a dictionary mapping state to list of universities in that state
by_state = {}
for uni in universities:
    state = uni["state-province"]
    if state not in by_state:
        by_state[state] = []
    by_state[state].append(uni["name"])

# Print the states that have more than one technology university
print("States with multiple technology universities:\n")
for state, unis in by_state.items():
    if len(unis) > 1 and state is not None:
        print(f"  {state}: {len(unis)} universities")
        for name in unis:
            print(f"    - {name}")
````

## Recap

| Concept | What it does | Example |
|---|---|---|
| **List** | Stores ordered items | `[88, 92, 75]` |
| **Dictionary** | Stores key-value pairs | `{"name": "Maria", "gpa": 3.7}` |
| **for loop** | Processes each item one at a time | `for x in my_list:` |
| **Append** | Adds an item to a list | `my_list.append(99)` |
| **API** | Fetches real data from the internet | `urllib.request.urlopen(url)` |

You now know how to:
- Create lists and dictionaries
- Use loops to process, filter, and summarize data
- Pull real data from APIs and work with it
- Combine lists and dictionaries to solve complex problems

These skills are the foundation of data science, web development, and nearly every Python project you'll encounter.
