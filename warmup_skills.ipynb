## Topic 1: `enumerate` — looping with an index

Normally a `for` loop gives you the **value**. `enumerate` gives you both the **index (position)** and the **value** at the same time.

```python
for index, word in enumerate(words):
    print(index, word)
```

You'll need this whenever you want to **change** something inside the list — because `word` is just a copy, but `words[index]` reaches into the real list.

````python
# ── Me ────────────────────────────────────────────────
# enumerate gives us both the position and the value

words = ['Four', 'score', 'and', 'seven', 'years']

for index, word in enumerate(words):
    print(index, word)
````

## Topic 2: Changing values in a list

You can't change a list item by changing the loop variable — it's just a copy.  
You have to write back into the list using the index:

```python
for index, name in enumerate(usernames):
    usernames[index] = name.lower()   # ← writes back into the real list
```

````python
# ── Me ────────────────────────────────────────────────
# Usernames were entered with inconsistent capitalization — fix them all

usernames = ['JohnDoe99', 'PRIYA_M', 'Carlos42']
print('Before:', usernames)

for index, name in enumerate(usernames):
    usernames[index] = name.lower()

print('After: ', usernames)
````

## Topic 3: Appending to a list

Sometimes you want to **build a new list** by picking only items that meet a condition.  
Start with an empty list, then use `.append()` inside a loop:

```python

for score in scores:
    if score >= 80:
        result.append(score)
```

````python
# ── Me ────────────────────────────────────────────────
# Collect only the scores that are 80 or above

grades = [72, 85, 91, 60, 78, 95, 55, 88]

high_scores = []
for grade in grades:
    if grade >= 80:
        high_scores.append(grade)

print('High scores:', high_scores)
````

## Topic 4: Working with a dictionary

A **dictionary** stores pairs: a key and a value.

```python
scoreboard = {}                                 # empty dictionary
scoreboard['Maya'] = 0                          # add a new key
scoreboard['Maya'] = scoreboard['Maya'] + 10    # update an existing value
```

To check whether a key already exists: `if 'Maya' in scoreboard`

````python
# ── Me ────────────────────────────────────────────────
# Build a scoreboard — every player starts at 0

players = ['Maya', 'Carlos', 'Priya']

scoreboard = {}
for player in players:
    scoreboard[player] = 0

print('Scoreboard:', scoreboard)

# Now Maya earns 10 points — update her value
scoreboard['Maya'] = scoreboard['Maya'] + 10
print('After Maya scores:', scoreboard)
````
