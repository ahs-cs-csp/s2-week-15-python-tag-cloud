## Step 1: Read the File

The first step is to **read words from a text file** and return them as a Python list.

If the file contains:
```
Four score and seven years ago
```
the function should return:
`['Four', 'score', 'and', 'seven', 'years', 'ago']`

### How it works
- Open the file with `open()`
- Read all the text as one big string with `.read()`
- Split by whitespace (spaces, tabs, newlines) with `.split()`
- Return the resulting list of words

````python
def read_file(file_name):
    # Read a text file and return a list of words split by whitespace.
    with open(file_name, 'r', encoding='utf-8') as file:
        text = file.read()
        words = text.split()
    return words
````

## Putting It All Together!

Now let's run the **complete pipeline** on `speech.txt` and generate the tag cloud.

The pipeline:
1. **Read** the file → raw list of words
2. **Remove special characters** → strip punctuation
3. **Lowercase** → normalise capitalisation
4. **Remove stop words** → filter insignificant words
5. **Count occurrences** → build `word_map`
6. **Remove rare words** → keep only the most meaningful

Then the **tag cloud** is drawn where:
- **Bigger** text = appeared more often
- Colours are random — just for fun!

Run the cell below to see your tag cloud!


````python
import math


def display_tag_cloud():
    # Spiral layout: pack words from centre outward, ~35% of smaller words rotated 90 degrees.
    if not word_map:
        print('word_map is empty — run all the cells above first!')
        return

    fig, ax = plt.subplots(figsize=(14, 9))
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.axis('off')
    fig.patch.set_facecolor('#fafafa')
    ax.set_facecolor('#fafafa')

    max_count   = max(word_map.values())
    min_count   = min(word_map.values())
    count_range = max_count - min_count if max_count > min_count else 1

    # Convert font-size (points) to axis data units.
    # Figure: 14 x 9 inches, axes: 0-100 x 0-100 data units.
    PX  = 100 / (14 * 72)
    PY  = 100 / (9  * 72)
    PAD = 1.5                # minimum gap (data units) between bounding boxes

    placed = []  # (cx, cy, w, h) of every word already drawn

    def estimate_box(word, fs, angle):
        # Approximate rendered size in data units.
        w = len(word) * fs * 0.68 * PX
        h = fs * 1.4 * PY
        return (h, w) if angle == 90 else (w, h)

    def overlaps(cx, cy, w, h):
        for px, py, pw, ph in placed:
            if (abs(cx - px) < (w + pw) / 2 + PAD and
                    abs(cy - py) < (h + ph) / 2 + PAD):
                return True
        return False

    def spiral(cx=50, cy=50):
        # Archimedean spiral from centre, squeezed to match the 14:9 aspect ratio.
        t = 0
        while t < 800:
            r = t * 0.28
            yield cx + r * math.cos(t), cy + r * math.sin(t) * (9 / 14)
            t += 0.1

    # Largest words first so they land near the centre.
    sorted_words = sorted(word_map.items(), key=lambda kv: kv[1], reverse=True)

    palette = [
        '#1a2e8c', '#b52020', '#1a7535', '#7a2a8a', '#b86010',
        '#0c6a7a', '#7a3a10', '#2a4a7a', '#6a1a50', '#1a4a20',
        '#0a3a6a', '#7a1010', '#0a5a4a', '#4a0a7a', '#6a4a00',
    ]

    n_placed = 0
    for i, (word, count) in enumerate(sorted_words):
        fs = 10 + (count - min_count) / count_range * 50

        # Top-frequency words stay horizontal; ~35% of the rest go vertical.
        if count >= max_count * 0.65:
            angle = 0
        else:
            angle = 90 if random.random() < 0.35 else 0

        w, h   = estimate_box(word, fs, angle)
        margin = 1.5

        for sx, sy in spiral():
            if sx - w / 2 < margin or sx + w / 2 > 100 - margin:
                continue
            if sy - h / 2 < margin or sy + h / 2 > 100 - margin:
                continue
            if not overlaps(sx, sy, w, h):
                color  = palette[i % len(palette)]
                weight = 'bold' if count > max_count * 0.5 else 'normal'
                ax.text(sx, sy, word, fontsize=fs, color=color,
                        ha='center', va='center', fontweight=weight,
                        rotation=angle)
                placed.append((sx, sy, w, h))
                n_placed += 1
                break

    ax.set_title('Tag Cloud', fontsize=18, fontweight='bold', pad=12, color='#333')
    plt.tight_layout()
    plt.show()
    print(f'Tag cloud: {n_placed} of {len(word_map)} words placed.')


def process_words():
    # Run the full pipeline then show the spiral tag cloud.
    global word_map
    word_map = {}

    words = read_file('speech.txt')
    print(f'Step 1 — Read {len(words)} words from the file')

    ...
    print(f'Step 2 — Removed special characters')

    ...
    print(f'Step 3 — Converted to lowercase')

    ...
    print(f'Step 4 — Removed insignificant words  ({len(words)} words remain)')

    ...
    print(f'Step 5 — Counted words: {len(word_map)} unique words found')

    ...
    print(f'Step 6 — Removed rare words: {len(word_map)} words remain')

    print()
    print('word_map:', word_map)
    print()

    display_tag_cloud()


process_words()
````
